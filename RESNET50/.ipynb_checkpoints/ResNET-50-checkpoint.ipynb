```python
# Imports
import tensorflow as tf
from tensorflow.keras.applications.resnet50 import ResNet50
```

```python
# ResNET implementation
# We use already trained ResNet-50 model (ImageNET dataset)
# However, we change last layer and train on our own dataset
def impl_ResNet50(output_length):
    model = ResNet50(input_shape=(224, 224, 3), include_top=False, weights='imagenet')
    X = model.output
    outputs = tf.keras.layers.Dense(output_length, activation='softmax')(X)
    resNet50 = tf.keras.Model(inputs=model.input, outputs=outputs)
    resNet50.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return resNet50
```

```python
# Load implemented model
model = impl_ResNet50(300)
model.summary()
```
